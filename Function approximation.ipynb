```python
import math

import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy import linalg
```

```python
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
```

```python
plt.plot([x for x in range(1, 16)], [f(x) for x in range(1, 16)])
```

```python
A = np.array([[1, 1], [1, 15]])
B = np.array([f(1), f(15)])

answer = scipy.linalg.solve(A, B)
answer
```

```python
plt.plot([1, 15], [answer[0], 15*answer[1]])
```

```python
A = np.array([[1, 1, 1], [1, 8, 8**2], [1, 15, 15**2]])
B = np.array([f(1), f(8), f(15)])
answer = scipy.linalg.solve(A, B)
answer
```

```python
plt.plot([1, 8, 15], [answer[0], 8*answer[1], 15*answer[2]])
```

```python
A = np.array([[1**x for x in range(4)], [4**x for x in range(4)], [10**x for x in range(4)], [15**x for x in range(4)]])
B = np.array([f(1), f(4), f(10), f(15)])
answer = scipy.linalg.solve(A, B)
answer
```

```python
plt.plot([1, 4, 10, 15], [answer[0], 4*answer[1], 10*answer[2], 15*answer[3]])
```

```python
answer = np.around(answer, decimals = 2)
answer
```

```python
answerFile = open('function_approximationanswer_for_similarity.txt', 'w')

for i in answer:
    answerFile.write(str(i))
    answerFile.write(' ')
    print(i)
```
